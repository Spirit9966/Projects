```python
urls = []
```

```python
with open('tableList.tsv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file, delimiter='\t')
    headers = ['Product Name', 'Price', 'Shipping', 'Number of Watcher', 'Condition', 'Image URL', 'Product URL']
    writer.writerow(headers)
    
    for product in products:
        
        name = product.find('div',class_='s-item__title').get_text()
        product_url = product.find('a').get('href')
        
        urls.append(product_url)

        price = product.find('span',class_='s-item__price').get_text()  

        shipping = product.find('span',class_='s-item__shipping s-item__logisticsCost')

        if shipping:
            shipping_cost =  shipping.get_text()
        else:
            shipping_cost = "No shipping info found"
    
        watchers = product.find('span', class_='s-item__dynamic s-item__watchCountTotal')
            
        if watchers:
            watch = watchers.get_text()
        else:
            watch = "No watchers info"
      
        condition = product.find('div',class_='s-item__subtitle').get_text()
        
        image_url = product.find('img')['src']
        
        row = [name,price,shipping,watch,condition,image_url,product_url ]
    
        writer.writerow(row)
```

```python
hdr = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.11 (KHTML, like Gecko) Chrome/23.0.1271.64 Safari/537.11',
       'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
       'Accept-Charset': 'ISO-8859-1,utf-8;q=0.7,*;q=0.3',
       'Accept-Encoding': 'none',
       'Accept-Language': 'en-US,en;q=0.8',
       'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
       'Connection': 'keep-alive'}



def scrapper(link):
    
    global df
    
    page = requests.get(link,headers=hdr)
        
    webpage = BeautifulSoup(page.text,'html')
    
    
    #Initialise an empty dict to store block 1 columns later on
    block_1_dict = {}
    
    #From the entire HTML, extract that information that is relevant to us
    block_1 = webpage.find('div',id='LeftSummaryPanel')


    #Get the name of the product
    name = block_1.find('span',class_='ux-textspans ux-textspans--BOLD').get_text()
    block_1_dict['Product Name'] = name

    #Get the returns information
    returns_info = block_1.find('div',class_='vim x-returns-minview')

    returns = returns_info.find('div','ux-labels-values__values col-9').get_text()
    
    substring = 'See details- for more information about returns'

    if returns.find(substring) != -1:
        block_1_dict['Returns'] = returns[0:returns.find(substring)]
    else:
        block_1_dict['Returns'] = returns


    #Get the payment modes available
    payment_methods = block_1.find_all('span',role="img")

    #Initiate an empty string first
    pay = ''

    for p in payment_methods:
        if p.get("title") != None:
            pay = pay + ',' + p.get("title")
        else:
            pass

    block_1_dict['Payment Modes'] = pay.lstrip(',')

    #Get the starting bid and the price - Not every product has a starting bid or a price.

    bid_or_price = block_1.find_all('div',class_='x-buybox__section')

    for price in bid_or_price:
        if price.find('div',class_='x-label'):
            label = price.find('div',class_='x-label').get_text().replace(":","")

        if price.find('div',class_='x-price-primary'):
            value = price.find('div',class_='x-price-primary').get_text()
            block_1_dict[label] = value

    srp = block_1.find_all('div',class_='ux-layout-section__row')

    for row in srp:
        label = row.find('div',class_='ux-labels-values__labels-content').get_text()
        content = row.find('div',class_='ux-labels-values__values-content').get_text()
        
        if label == 'Shipping:':
            substring = 'See detailsfor shipping'
            if content.find(substring) != -1:
                block_1_dict['Shipping cost'] = content[0:content.find(substring)].replace("\xa0",".")
            else:
                block_1_dict[label] = content
    
        elif label == 'Delivery:':
            substring = 'Estimated delivery dates - opens in a new window'
            if content.find(substring) != -1:
                block_1_dict['Estimated delivery date'] = content[0:content.find(substring)]
            else:
                block_1_dict['Estimated delivery date'] = content
        
        else:
            pass
    

        
    ##################################################################################   

    ## Block - II

    block_2_dict = {}
    
    headers = []
    values = []

    #### "About This Item" Section

    #The following is going to find two results: One for "About This Item" and the other for "Shipping, Returns & Payments"
    block_2 = webpage.find_all('div',class_='vi-VR-tabCnt')

    #Access the "About this item" section
    rows = block_2[0].find_all('div',class_='ux-layout-section__row')


    #Get the eBay number
    headers.append(rows[1].get_text().split(":")[0])
    values.append(rows[1].get_text().split(":")[1])

    for row in rows:
        labels =  row.find_all('div',class_='ux-labels-values__labels')
        content = row.find_all('div',class_='ux-labels-values__values')
        
        for l in labels:
            headers.append(l.get_text().replace(":",""))

        for c in content:
            substring = 'Read moreabout the condition'
            field_content = c.get_text()
        
            if field_content.rfind(substring) != -1:
                field_content = field_content[field_content.rfind(substring)+28:]
                values.append(field_content.replace("opens in a new window or tab",""))
            else:
                values.append(field_content)

        for i in range(len(headers)):
            block_2_dict[headers[i]] = values[i]

        #Clear off the lists
        headers.clear()
        values.clear()



    #Access the "Shipping, returns and payments" section
    srp = block_2[1].find('div',{'id':'vi-ship-maincntr'})


    #Get the "Item Location" and "Ships To" information
    info = srp.find_all('div',class_='ux-layout-section__row')

    for i in info:
        if i.find('div',class_='ux-labels-values__labels col-6'):
            dict_key = i.find('div',class_='ux-labels-values__labels col-6').get_text().replace("\xa0","").replace(":","")
        else:
            pass

        if i.find('div',class_='ux-labels-values__values col-6'):
            dict_val = i.find('div',class_='ux-labels-values__values col-6').get_text()
            block_2_dict[dict_key] = dict_val
        else:
            pass
        
    ##################################################################################   
    
    #Combine the two dictionaries into one
    final_dict = {**block_1_dict,**block_2_dict}

    required_columns = ['Brand','Condition','eBay item number','Material','Type','Item location','Ships to','Product Name','Price','Starting bid','Payment Modes','Returns','Estimated delivery date','Shipping cost','Color']

    for col in required_columns:
        if col not in final_dict.keys():
            final_dict[col] = 'Not Found'

    #Finally, filter the dictionary to only include the required key-value pairs
    final_dict = {key:final_dict[key] for key in required_columns}   

    df = df.append(final_dict,ignore_index=True)
    
```

```python
df = pd.DataFrame()


for url in urls:
    scrapper(url)
```

```python
#Re-arrange the columns

df = df[['Product Name','Price','Starting bid','Payment Modes','Returns','Brand','Material','Color','Type','Condition','eBay item number','Ships to','Item location','Shipping cost','Estimated delivery date']]    

#Save the output as a tab separated file
df.to_csv("tableDetails.tsv",sep='\t')


df
```
